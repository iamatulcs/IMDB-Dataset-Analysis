# **Q25. Do movies with more actors (ensemble cast) earn more?**

```python
df_exploded = df.assign(stars_name=df["stars_name"].str.split(",")).explode("stars_name")
```

```python
actor_count = df_exploded.groupby("name")["stars_name"].count().reset_index()
actor_count.columns = ["name", "actor_count"]
```

```python
merged = df_exploded[["name", "gross_income"]].drop_duplicates().merge(actor_count, on="name")
```

```python
correlation = merged["actor_count"].corr(merged["gross_income"])
print(f"Correlation between number of actors and gross: {correlation:.2f}")
```

# **Q26. Which Director has the highest "Hit Rate"? (Percentage of movies rated > 7.5)**

```python
df["is_hit"] = df["rating"] > 7.5
```

```python
director_stats = df.groupby("directors_name").agg(
    total_movies=("name", "count"),
    hit_movies=("is_hit", "sum")
)
```

```python
director_stats["hit_rate"] = (
    director_stats["hit_movies"] / director_stats["total_movies"]
)
```

```python
filtered = director_stats[director_stats["total_movies"] >= 3]
```

```python
top_director = filtered.sort_values("hit_rate", ascending=False).head(1)
```

```python
print("Director with Highest Hit Rate:")
print(top_director)
```
